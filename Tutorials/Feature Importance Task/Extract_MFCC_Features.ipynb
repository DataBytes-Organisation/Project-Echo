```python
import os
import librosa
import numpy as np

# 设置你的数据路径
audio_path = r"D:\project echo\Project-Echo-main\src\Prototypes\data\data_files"
output_path = r"D:\project echo\Project-Echo-main\src\Prototypes\data\mfcc_features"

# 创建输出目录（如果不存在）
os.makedirs(output_path, exist_ok=True)

for folder in os.listdir(audio_path):
    folder_path = os.path.join(audio_path, folder)
    if os.path.isdir(folder_path):
        save_folder = os.path.join(output_path, folder)
        os.makedirs(save_folder, exist_ok=True)

        for file in os.listdir(folder_path):
            if file.endswith(".mp3") or file.endswith(".wav"):
                file_path = os.path.join(folder_path, file)
                try:
                    y, sr = librosa.load(file_path, sr=None)
                    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
                    mfcc_file = os.path.splitext(file)[0] + ".npy"
                    np.save(os.path.join(save_folder, mfcc_file), mfcc)
                    print(f"Saved MFCC for {file} in {save_folder}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")
```
