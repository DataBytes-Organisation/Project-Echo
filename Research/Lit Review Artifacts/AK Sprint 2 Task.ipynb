```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Full path to the selected audio file
file_path = os.path.join(data_path, selected_file)

# Load the audio file
audio, sample_rate = librosa.load(file_path, sr=None)

# Display basic information
print(f"Loaded '{selected_file}' with a sample rate of {sample_rate} Hz")
```

```python
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio, sr=sample_rate)
plt.title(f"Waveform of {selected_file}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
from IPython.display import Audio

# Play the audio
Audio(data=audio, rate=sample_rate)
```

```python
import ipywidgets as widgets
from IPython.display import display
import os

# Dropdown menu to select a file
dropdown = widgets.Dropdown(
    options=audio_files,
    description='Select File:',
    value=audio_files[0]  # Default to the first file
)

def on_dropdown_change(change):
    selected_file = change['new']  # Get the new selection
    file_path = os.path.join(data_path, selected_file)
    print(f"Selected file: {selected_file}")
    print(f"File path: {file_path}")

# Attach a handler to capture changes in selection
dropdown.observe(on_dropdown_change, names='value')

# Display the dropdown
display(dropdown)
```
