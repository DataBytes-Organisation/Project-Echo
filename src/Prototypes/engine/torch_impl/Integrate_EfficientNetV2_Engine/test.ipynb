## Validate TFLite-ReaL Engine intergration with fake MQTT

```python
import json
import base64

os.chdir(
    r"C:\Deakin\ProjectEcho\Project-Echo\src\Prototypes\engine\torch_impl"
)

from light_echo_engine_efficientnetv2_tflite import EchoEngine

engine = EchoEngine()

AUDIO_PATH = r"C:\Deakin\ProjectEcho\Project-Echo\src\Prototypes\data\data_files\Acanthiza chrysorrhoa\region_3.650-4.900.mp3"

with open(AUDIO_PATH, "rb") as f:
    audio_bytes = f.read()

audio_base64 = base64.b64encode(audio_bytes).decode("utf-8")

test_audio_event = {
    "timestamp": "2026-04-30T10:00:00",
    "audioFile": "EfficientNetV2_TFLite_Mode",
    "audioClip": audio_base64,
    "sensorId": "test_sensor_001",
    "microphoneLLA": [-37.8136, 144.9631, 0],
    "animalEstLLA": [-37.8136, 144.9631, 0],
    "animalTrueLLA": [-37.8136, 144.9631, 0],
    "animalLLAUncertainty": 0
}


class FakeMQTTMessage:
    def __init__(self, payload):
        self.payload = payload


fake_msg = FakeMQTTMessage(
    payload=json.dumps(test_audio_event).encode("utf-8")
)

engine.on_message(None, None, fake_msg)
```
