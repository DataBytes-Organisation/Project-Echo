```python
import librosa
import librosa.display

import matplotlib.pyplot as plt
import numpy as np

# Load the audio file (replace 'audio_file.wav' with your audio file)
audio_file = 'cat_3.wav'
y, sr = librosa.load(audio_file)

# Compute the Mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

# Convert to decibels
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Plot the Mel spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(mel_spectrogram_db, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')

# Save the Mel spectrogram as a PNG file (replace 'mel_spectrogram.png' with your desired file name)
plt.savefig('ms1.png', bbox_inches='tight', pad_inches=0)

plt.show()
```
