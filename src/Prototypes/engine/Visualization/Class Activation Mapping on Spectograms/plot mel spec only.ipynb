```python
# AUDIO_CLIP_DURATION = 5 seconds 
AUDIO_NFFT =  2048
AUDIO_WINDOW = None
AUDIO_STRIDE = 200
AUDIO_SAMPLE_RATE = 48000
AUDIO_MELS = 260
AUDIO_FMIN = 20
AUDIO_FMAX =13000
AUDIO_TOP_DB = 80
```

```python
# plot a single mel spec

y, sr = librosa.load(r"library_where_your_audio_is_stored")
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr,
                                                n_mels=AUDIO_MELS, hop_length=AUDIO_STRIDE, fmax=AUDIO_FMAX)

fig = plt.figure(figsize = (5,5), frameon=False)
plt.tight_layout(pad=0)
img = librosa.display.specshow(librosa.power_to_db(
    mel_spectrogram, ref=np.max),hop_length = 128, fmax=8000)
plt.axis("off")

### change the output directory and filename
plt.savefig("testing3.png", bbox_inches='tight', pad_inches=0)
```
