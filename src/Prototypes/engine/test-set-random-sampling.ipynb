## Testing Dataset Creation

> This script will randomly sample 3 audio files from each species folder into the target test directory.

```python
input_bucket_dir = 'D:\\Projects\\project-echo-data\\bucket-1-updated\\Otways Species'
output_bucket_dir = 'D:\\Projects\\project-echo-data\\bucket-1-updated-test'
```

```python
import os
import shutil
import random
```

```python
def copy_directory_structure(source_dir, dest_dir):
    # Walk through the source directory
    for root, _, files in os.walk(source_dir):
        # Calculate the relative path to the current directory
        relative_dir = os.path.relpath(root, source_dir)
        # Construct the corresponding directory in the destination
        dest_path = os.path.join(dest_dir, relative_dir)
        
        # Create the directory in the destination if it doesn't exist
        os.makedirs(dest_path, exist_ok=True)

copy_directory_structure(input_bucket_dir, output_bucket_dir)
```

```python
dir = os.listdir(input_bucket_dir)
dir = [subdir for subdir in dir if not subdir.startswith('.')]
print(dir)
```

```python
def move_random_files(source_dir, destination_dir, num_files_to_move):
    # List all files in the source directory and its sub-directories
    all_files = []
    for root, _, files in os.walk(source_dir):
        for file in files:
            all_files.append(os.path.join(root, file))

    # Randomly select num_files_to_move files from the list
    try:
        random_files = random.sample(all_files, num_files_to_move)
    except:
        print(f'Error: {num_files_to_move} were not found in {source_dir}.')
        try:
            random_files = random.sample(all_files, len(os.listdir(source_dir)))
        except:
            print(f'Error: {len(os.listdir(source_dir))} were not found in {source_dir}.')
    
    # Move the selected files to the destination directory
    for file_path in random_files:
        # Calculate the relative path from the source directory
        relative_path = os.path.relpath(file_path, source_dir)
        # Construct the destination path
        dest_path = os.path.join(destination_dir, relative_path)
        # Create the destination directory if it doesn't exist
        os.makedirs(os.path.dirname(dest_path), exist_ok=True)
        # Move the file
        shutil.move(file_path, dest_path)
        print(f"Moved: {file_path} to {dest_path}")

for subdir in dir:
    move_random_files(os.path.join(input_bucket_dir, subdir), os.path.join(output_bucket_dir, subdir), 3)
```
