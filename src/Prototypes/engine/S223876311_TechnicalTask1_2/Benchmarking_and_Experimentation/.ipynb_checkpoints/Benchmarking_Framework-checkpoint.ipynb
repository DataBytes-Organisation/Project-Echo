## 5. Experiment Runner Functions

These functions handle the execution of experiments and report generation.

```python
# --- Imports used by the runner ---
import importlib, os, re
from IPython.display import clear_output
import config.system_config as system_config     # module, so we can reload
from config.system_config import SC              # dict (we'll mutate this)
from config.experiment_configs import EXPERIMENTS
from utils.optimised_engine_pipeline import train_model

def _sync_dirs_to_memory(new_data_dir: str, new_cache_dir: str, new_output_dir: str):
    """Update SC in-memory so downstream code sees new paths immediately."""
    SC["AUDIO_DATA_DIRECTORY"] = new_data_dir
    SC["CACHE_DIRECTORY"] = new_cache_dir
    SC["OUTPUT_DIRECTORY"] = new_output_dir

def _try_patch_system_config_file(new_data_dir: str, new_cache_dir: str, new_output_dir: str):
    """Best-effort: patch system_config.py on disk, then reload the module."""
    cfg_path = os.path.join('config', 'system_config.py')
    try:
        with open(cfg_path, 'r') as f:
            txt = f.read()
        txt2 = re.sub(r"('AUDIO_DATA_DIRECTORY':\s*r\")[^\"]*(\")", rf"\1{new_data_dir}\2", txt)
        txt2 = re.sub(r"('CACHE_DIRECTORY':\s*r\")[^\"]*(\")", rf"\1{new_cache_dir}\2", txt2)
        txt2 = re.sub(r"('OUTPUT_DIRECTORY':\s*r\")[^\"]*(\")", rf"\1{new_output_dir}\2", txt2)
        if txt2 != txt:
            with open(cfg_path, 'w') as f:
                f.write(txt2)
        # Reload the module properly
        importlib.reload(system_config)
    except Exception as e:
        print(f"(non-fatal) Could not update/reload {cfg_path}: {e}")

def run_selected_experiments(b):
    output_area.clear_output(wait=True)
    with output_area:
        print("Starting experiment run...")

        # 1) Get dirs from widgets and update SC in-memory (and try to persist to file)
        new_data_dir  = data_dir_widget.value.strip()
        new_cache_dir = cache_dir_widget.value.strip()
        new_out_dir   = output_dir_widget.value.strip()
        _sync_dirs_to_memory(new_data_dir, new_cache_dir, new_out_dir)
        _try_patch_system_config_file(new_data_dir, new_cache_dir, new_out_dir)
        print(f"Using DATA_DIR={SC['AUDIO_DATA_DIRECTORY']}")
        print(f"Using CACHE_DIR={SC['CACHE_DIRECTORY']}")
        print(f"Using OUTPUT_DIR={SC['OUTPUT_DIRECTORY']}")

        # 2) Which experiments?
        selected_names = list(experiment_widget.value)
        if not selected_names:
            print("No experiment selected. Please select at least one.")
            return

        # 3) Run each selected experiment
        for name in selected_names:
            exp_config = next((e for e in EXPERIMENTS if e["name"] == name), None)
            if exp_config is None:
                print(f"Experiment '{name}' not found. Skipping.")
                continue

            # IMPORTANT: set per-experiment augmentation FLAGS **inside** the loop
            SC['IMAGE_AUGMENTATION'] = exp_config.get('image_augmentation', 'none')
            SC['AUDIO_AUGMENTATION'] = exp_config.get('audio_augmentation', 'none')
            print(f"\nRunning experiment: {exp_config['name']}")
            print(f"  model={exp_config['model']}")
            print(f"  AUDIO_AUGMENTATION={SC['AUDIO_AUGMENTATION']} | IMAGE_AUGMENTATION={SC['IMAGE_AUGMENTATION']}")

            # Call trainer
            model, history = train_model(
                model_name=exp_config['model'],
                epochs=exp_config.get('epochs'),
                batch_size=exp_config.get('batch_size')
            )
            print(f"Training completed: {exp_config['name']}")
            if model:
                model.summary(print_fn=lambda s: print(s))
            print("-" * 60)

def run_all_experiments(b):
    # Select all, then reuse the same flow
    experiment_widget.value = tuple([e["name"] for e in EXPERIMENTS])
    run_selected_experiments(b)

# Wire up buttons
run_selected_button.on_click(run_selected_experiments)
run_all_button.on_click(run_all_experiments)
# generate_report_button.on_click(your_generate_report_fn)  # optional
```

## 6. View Previous Results (From here down, notebook is under development)

If you've already run experiments, you can view and analyse the results here.

```python
# Under development
def load_results(output_dir=OUTPUT_DIR):
    # Check if results directory exists
    if not os.path.exists(output_dir):
        print(f"Results directory does not exist: {output_dir}")
        return None
    
    # Look for comparison report CSV
    csv_files = [f for f in os.listdir(output_dir) if f.startswith("comparison_results_") and f.endswith(".csv")]
    
    if not csv_files:
        print("No comparison results found. Run experiments or generate a report first.")
        return None
    
    # Load the latest CSV file
    latest_csv = max(csv_files)
    csv_path = os.path.join(output_dir, latest_csv)
    results_df = pd.read_csv(csv_path)
    
    print(f"Loaded results from: {csv_path}")
    return results_df

# Load and display results if available
results_df = load_results()
if results_df is not None:
    display(results_df)
```
